```python
percent_paid = ((loan.shape[0] - loan[loan['past_due_days'] == 0].shape[0]) / (loan.shape[0])) * 100
print(percent_paid, '% of people paid after time')
```

```python
labels = ['After Due Date', 'Before Due Date']
size = [percent_paid, 100 - percent_paid]
explode = (0, 0.1)
```

```python
fig1, ax1 = plt.subplots(figsize = (8, 4))
plt.suptitle('Percentage of People Paid Before or After Due Date', fontsize = 15, fontweight = 'bold')
plt.pie(x = size, explode = explode, autopct = '%1.1f%%', shadow = True, startangle = 90, colors = ['#577399', '#BDD5EA'])
ax1.axis('equal')
plt.legend(labels, bbox_to_anchor = (0.2, 1))
```

```python
fig = plt.figure(figsize=(15,5))
ax = sns.countplot(x="effective_date", hue="loan_status", data=loan ,palette='pastel' ,edgecolor=sns.color_palette("dark", 3))
ax.set_title('Loan date')
ax.tick_params('x', labelrotation = 45)
ax.legend(loc='upper right')
for t in ax.patches:
    if (np.isnan(float(t.get_height()))):
        ax.annotate(0, (t.get_x(), 0))
    else:
        ax.annotate(str(format(int(t.get_height()), ',d')), (t.get_x(), t.get_height()*1.01))
plt.show();
```

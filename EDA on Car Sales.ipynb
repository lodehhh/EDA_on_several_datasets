```python
missing_values = car.isnull().sum()
missing_values
```

```python
total_cells = np.product(car.shape)
total_cells
```

```python
total_missing_values = missing_values.sum()
print('Total Missing Values are', total_missing_values)
```

```python
percent_missing_values = (total_missing_values / total_cells) *100
print('Percentage of Total Missing Values are', percent_missing_values)
#Percentage of missing values is very small, we can just drop all the values within the rows of missing values.
#Instead I will replace the missing values with mean of each columns so the analysis can be more precise
```

```python
car['Resale_values'].fillna(car['Resale_values'].mean(), inplace =True)
car["Price_in_thousands"].fillna(car["Price_in_thousands"].mean(), inplace=True)
car["Engine_size"].fillna(car["Engine_size"].mean(), inplace=True)
car["Horsepower"].fillna(car["Horsepower"].mean(), inplace=True)
car["Wheelbase"].fillna(car["Wheelbase"].mean(), inplace=True)
car["Width"].fillna(car["Width"].mean(), inplace=True)
car["Length"].fillna(car["Length"].mean(), inplace=True)
car["Curb_weight"].fillna(car["Curb_weight"].mean(), inplace=True)
car["Fuel_capacity"].fillna(car["Fuel_capacity"].mean(), inplace=True)
car["Fuel_efficiency"].fillna(car["Fuel_efficiency"].mean(), inplace=True)
car["Power_perf_factor"].fillna(car["Power_perf_factor"].mean(), inplace=True)
```

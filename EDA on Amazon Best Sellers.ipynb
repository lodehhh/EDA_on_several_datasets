```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
bar = book['Author'].value_counts().sort_values(ascending = False).head(5)
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = bar, y = bar.index, palette = ['#495867', '#577399', '#BDD5EA', '#F7F7FF', '#FE5F55'])
plt.xlabel('Counts', fontsize = 15)
plt.ylabel('Author', fontsize = 15)
plt.title('Top 5 Authors with Most Books (2009 -2019)', fontsize = 20, fontweight = 'bold')
```

```python
pie = book['Genre'].value_counts()
labels = ['Non Fiction', 'Fiction']
```

```python
plt.figure(figsize = (10, 6))
plt.pie(pie, autopct = '%0.f%%', shadow = True, colors = ['#577399', '#BDD5EA'])
plt.legend(labels = labels, bbox_to_anchor = (0,1), fontsize = 15)
plt.title('Genres Percantage (2009 - 2019)', fontsize = 18, fontweight = 'bold')
```
